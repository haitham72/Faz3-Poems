```python
import pandas as pd
import re

def remove_arabic_diacritics(text):
    """Remove all Arabic diacritics (tashkeel)"""
    if pd.isna(text) or text == "":
        return text
    # Remove Arabic diacritics
    arabic_diacritics = re.compile(r'[\u064B-\u065F\u0670]')
    return arabic_diacritics.sub('', str(text))

def clean_text(text):
    """Remove diacritics, extra dashes, dots at start/end, normalize whitespace"""
    if pd.isna(text) or text == "":
        return text
    
    text = str(text)
    
    # Remove Arabic diacritics
    text = remove_arabic_diacritics(text)
    
    # Remove leading/trailing dots and dashes
    text = text.strip('.-_ \t')
    text = text.replace('"', '')

    
    return text.strip()

def clean_content_with_separator(text):
    """Clean content - preserve formatting, only remove diacritics"""
    if pd.isna(text) or text == "":
        return text
    
    text = str(text)
    
    # ONLY remove diacritics, preserve everything else
    text = remove_arabic_diacritics(text)
    
    # Normalize line endings only
    text = text.replace('\r\n', '\n').replace('\r', '\n')
    
    return text

# Load your Excel file
df = pd.read_excel('POEMS - Faz3 - tagged.xlsx')

# Print column names to see actual names
print("Column names in your file:")
print(df.columns.tolist())
print()

# Find the actual column name for poem context (it might have different spacing/capitalization)
poem_context_col = None
for col in df.columns:
    if 'poem' in col.lower() and 'context' in col.lower():
        poem_context_col = col
        break

if poem_context_col:
    print(f"Found poem context column: '{poem_context_col}'")
    df[poem_context_col] = df[poem_context_col].apply(clean_text)
else:
    print("⚠️  Could not find poem context column")

# Clean content (preserve summary ----- chunk structure)
if 'content' in df.columns:
    print("Cleaning content column...")
    df['content'] = df['content'].apply(clean_content_with_separator)
    
    # Verify content structure is preserved
    print("\n=== Sample content (first 300 chars) ===")
    print(repr(df.loc[0, 'content'][:300]))
else:
    print("⚠️  'content' column not found")

# Clean other text fields if they exist
for col in ['people', 'places', 'objects']:
    if col in df.columns:
        df[col] = df[col].apply(remove_arabic_diacritics)
        print(f"✅ Cleaned '{col}' column")


# Show first row to verify
print("\n=== First row sample ===")
if 'content' in df.columns:
    print(df.loc[0, 'content'][:200])
```
